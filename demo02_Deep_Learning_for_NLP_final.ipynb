## Q1 : Use Tokenizer from keras to process data and labels

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# processing data
texts = pre_processed_tweets_list
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
data = pad_sequences(sequences, maxlen=32)
vocab = tokenizer.word_index
vocab['<eos>'] = 0
```

## Q2: Cut this corpus in two parts one for the training and one for the evaluation 
>- (x_train and y_train ) should contains 4000 utterances
>- (x_val, y_val) should have the rest of the data
>- Feel free to shuffle data

```python
pd.concat([pd.DataFrame(data) , pd.DataFrame(labels)], axis = 1)
```

```python
from sklearn.utils import shuffle

data, labels = shuffle(data, labels)

x_train, y_train = data[:4000], labels[:4000]
x_val, y_val = data[4000:], labels[4000:]
```

# II. Train and evaluate the model

```python
from tp_sentiment_analysis import embedding

weights = embedding.load(vocab, 100, 'tp_sentiment_analysis/glove.twitter.27B.100d.filtered.txt')
pd.DataFrame(weights)
```

```python
vocab.keys()
```

```python
y_train[0]
```

```python
from keras.layers import Embedding, Input, GRU, Dense
from keras.models import Model
import tensorflow as tf
# 100-dim embeddings initialized with GloVe,
# over sequences of size 32, and not fine tuneable
embedding_layer = Embedding(len(vocab), 100, weights=[weights],input_length=32, trainable=False)
sequence_input = Input(shape=(32,), dtype='int32')
embedded_sequences = embedding_layer(sequence_input)
x = GRU(64)(embedded_sequences)
preds = Dense(labels.shape[1], activation='softmax')(x)
model = Model(sequence_input, preds)
model.compile(loss='categorical_crossentropy',optimizer='Nadam',\
              metrics=[tf.keras.metrics.Precision(),\
                       tf.keras.metrics.Recall() ,\
                       'acc',\
                       tf.keras.metrics.Accuracy()])
print(model.summary())
model.fit(x_train, y_train, validation_data=(x_val, y_val),epochs=2, batch_size=64, shuffle=True)
```

```python
loss, precision, recall,  acc, f1_measure = model.evaluate(x_val, y_val, batch_size=64)

print("Test loss :", f1_measure)
print("Test accuracy:", acc)
print("Test precision:", precision)
print("Test recall:", recall)
```

```python
from keras.layers import Conv1D, MaxPooling1D, Flatten

sequence_input = Input(shape=(32,), dtype='int32')
embedded_sequences = embedding_layer(sequence_input)
x = Conv1D(128, 3, activation='relu')(embedded_sequences)
x = MaxPooling1D(3)(x) # global max pooling
x = Flatten()(x)
preds = Dense(labels.shape[1], activation='softmax')(x)

model = Model(sequence_input, preds)
model.compile(loss='categorical_crossentropy',optimizer='Nadam', metrics=['acc'])
model.fit(x_train, y_train, validation_data=(x_val, y_val),epochs=10, batch_size=64, shuffle=True)
```
